#### Create some fake data to play around with 
   if csv file or most of the time would be excel data. pandas has method .read_excel and .read_csv, to help parse the excel data.

```python
target =pd.Series(np.random.randint(low = 20, high = 60, size=len(rowName)),index=rowName,name='targets')
d ={
    'v00r00':  pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
    'Unnamed : 41': pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
    '% Target.1': pd.Series(1,index=rowName),
    'v01r00' : pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
    '% Target.3': pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
    'Unnamed : 112': pd.Series(112,index=rowName),
    'v02r00': pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
    'v02r01': pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
    'v02r01a': pd.Series(abs(np.random.randn(len(rowName))),index=rowName),
   }
df = pd.DataFrame(d)
df
```

### Remove columns that are not needed like


```python
df.drop(df.columns[df.columns.str.contains('Unnamed',case=False)],axis=1,inplace=True)
df.drop(df.columns[df.columns.str.contains('% Target',case=False)],axis=1,inplace=True)
df.columns
```
